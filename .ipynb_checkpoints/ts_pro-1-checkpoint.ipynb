##### Data preprocessing

looking at data:

- 14 lakh data for 49 location each contains nearly 3k data from 2008 - 2017
- Evaporation, sunshine, cloud9,cloud3 contains null values.. delete them
- impute the remainings with mean
- target: Rain Today or Rain Tomorrow , map Yes=1, No=0
- remove categorical cols: winddir,etc.
- normalize the data.. as some contains max beyond outlier..
- as this is time-series and dates are continuous, we can drop the dates cols..
- don't shuffle the data..
- lag of 6 days.

```python
alb=df[df['Location']=='Albury'].copy()
```

```python
alb.info()
```

```python
alb.drop(['Date','Location','Evaporation','Sunshine','WindGustDir','Cloud9am','Cloud3pm'],inplace=True,axis=1)
```
