### Framework to approach Time-Series

5. Persistence Model : A simple baseline score to update the future models

7. ARIMA models : Simple univariate model
8. Model Validation : Validatation

```python
# have a validation set
split = len(df)-12
dataset,validation = df[:split]  , df[split:]
len(dataset),len(validation)
```

```python
# train/test split
X = dataset.values
X = X.astype('float32')
train_size = int(len(X) * 0.50)
train,test = X[0:train_size],X[train_size:]
```

```python
#make a persistance model
# for current trainx output is trainx+1
from sklearn.metrics import mean_squared_error
history = [x for x in train]
predictions = list()
for i in range(len(test)):
    yhat = history[-1] #one-lagged
    predictions.append(yhat)
    obs = test[i]
    history.append(obs)
    print(">Pridicted=%.3f, Expected=%.3f" % (yhat,obs))
rmse = np.math.sqrt(mean_squared_error(test,predictions))
print('RMSE:%.3f' % rmse) 
```

Persistent Model RMSE: 51.844

we need a Model that can lower this error..


```python
prd =np.array(predictions)
```

```python
notr=np.array([None for i in train]).reshape((-1,1))
```

```python
type(notr), notr.ndim, train.ndim
```

```python
plt.plot(train,label='train')
plt.plot([i for i in notr]+[x for x in prd],label='prediction')
plt.plot([i for i in notr] + [x for x in test],label='test')
plt.legend()
plt.grid(True)
```
